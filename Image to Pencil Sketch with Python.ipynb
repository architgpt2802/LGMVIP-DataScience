We need to read the image in RBG format and then convert it to a grayscale image. This will turn an image into a classic black and white photo. Then the next thing to do is invert the grayscale image also called negative image, this will be our inverted grayscale image. Inversion can be used to enhance details. Then we can finally create the pencil sketch by mixing the grayscale image with the inverted blurry image. This can be done by dividing the grayscale image by the inverted blurry image. Since images are just arrays, we can easily do this programmatically using the divide function from the cv2 library in Python.

```python
# Importing Libraries
import cv2
import matplotlib.pyplot as plt
```

#### Loading the Image

```python
#loading the image in the variable image
image = cv2.imread("C:\\Users\\ARCHIT\\Desktop\\LetsGrowMore\\img.jpg")#, cv2.IMREAD_GRAYSCALE)
image = cv2.resize(image, (640, 480))    #resizing the image
image = cv2.cvtColor(image, cv2.COLOR_BGR2RGB)    #converting to RGB as the image originally is in RGB color scheme
plt.imshow(image)
plt.show()
```

#### Converting Image into gray scaled image mode

```python
image1 = cv2.cvtColor(image, cv2.COLOR_RGB2GRAY)
#converting the image back to 3 channels image
gray_image = cv2.cvtColor(image1, cv2.COLOR_GRAY2BGR)
plt.imshow(gray_image)
plt.show()
```

#### Inverting the gray scaled image

```python
inverted_image = 255 - gray_image
plt.imshow(inverted_image)
plt.show()
```

#### Blurring the gray scaled image using GaussianBlur Function

```python
blurred = cv2.GaussianBlur(inverted_image, (21, 21), 0)
plt.imshow(blurred)
plt.show()
```

#### Inverting Blurred Image to Pencil Image

```python
inverted_blurred = 255 - blurred
plt.imshow(inverted_blurred)
plt.show()
pencil_sketch = cv2.divide(gray_image, inverted_blurred, scale=246.0)
plt.imshow(pencil_sketch)
plt.show()
```

#### Adjusting the contrast of the pencil sketch

```python
alpha = 0.9 # Contrast control
beta = 30 # Brightness control (0-100)

contrast_adjust = cv2.convertScaleAbs(pencil_sketch, alpha=alpha, beta=beta)

plt.imshow(image)
plt.title("Original Image")
plt.show()
plt.imshow(contrast_adjust)
plt.title("Pencil Sketch")
plt.show()
```

The original image is converted into an image looking like a pencil drawn sketch.
